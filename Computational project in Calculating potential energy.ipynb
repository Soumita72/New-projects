```python
import numpy as np
sigma = float(input("Give the value of sigma "))
lamda = int(input("Give the value of lamda "))
epsilon = int(input("Give the value of epsilon "))
dr=float(input("Give the value of dr "))
neu= int(input("Give the value of neu "))
choice = int(input("Give the choice "))
rlist=[]
Vlist=[]

# Calculation of potential energy by square well potential model
if choice==1:
    for r in np.arange(3,15,dr):
        if r<3.4:
            V=float('inf')
        elif (sigma<= r <(sigma*lamda)):
            V= -1*epsilon
        else:
            V=0
        rlist.append(r)
        Vlist.append(V)
        
    print(rlist)
    print(Vlist)
    
#Calculation of potential energy by soft sphere model
elif choice==2:
    for r in np.arange(3,15,dr):
        V=1*((epsilon*(sigma/r))**neu)
        rlist.append(r)
        Vlist.append(V)
        
    print(rlist)
    print(Vlist)
    
#Calculation potential energy by Lenard Jones model    
else:
    for r in np.arange(3,15,dr):
        V=4*epsilon*(((sigma/r)**12)-((sigma/r)**6))
        rlist.append(r)
        Vlist.append(V)
        
    print(rlist)
    print(Vlist)
    
import matplotlib.pyplot as plt
x=np.array(rlist)
y=np.array(Vlist)
plt.plot(x,y)
plt.title('Potential vs internuclear distance plot')
plt.xlabel("internuclear distance")
plt.ylabel("Potential energy")
plt.show()
```
